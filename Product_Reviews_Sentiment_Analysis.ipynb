```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('/content/sample_data/Amazon-Product-Reviews-Sentiment-Analysis-in-Python-Dataset.csv')
data.head()
```

```python
data.info()
```

```python
data.loc[data['Sentiment'] <= 3, 'Sentiment'] = 0

data.loc[data['Sentiment'] > 3, 'Sentiment'] = 1
```

```python
stp_words = set(stopwords.words('english'))

def clean_review(review):
    if isinstance(review, str):
        return " ".join(word for word in review.split() if word.lower() not in stp_words)
    else:
        return ""  # or np.nan if you prefer to keep it null

data['Review'] = data['Review'].apply(clean_review)
```

```python
data.head()
```

```python
data['Sentiment'].value_counts()
```

```python
consolidated=' '.join(word for word in data['Review'][data['Sentiment']==0].astype(str))
wordCloud=WordCloud(width=1600,height=800,random_state=21,max_font_size=110)
plt.figure(figsize=(15,10))
plt.imshow(wordCloud.generate(consolidated),interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
consolidated=' '.join(word for word in data['Review'][data['Sentiment']==1].astype(str))
wordCloud=WordCloud(width=1600,height=800,random_state=21,max_font_size=110)
plt.figure(figsize=(15,10))
plt.imshow(wordCloud.generate(consolidated),interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
cv = TfidfVectorizer(max_features=2500)
x = cv.fit_transform(data['Review']).toarray()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test  = train_test_split(x, data['Sentiment'], test_size = 0.25, random_state= 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()

model.fit(x_train, y_train)

pred = model.predict(x_test)

print(accuracy_score(y_test, pred))
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [False, True])

cm_display.plot()

plt.show()
```
